# TCLab_Test

```python
import tclab
import time
```

# Connect to TCLab

```python
lab = tclab.TCLab()

lab.Q1(50)
lab.Q2(50)

# Load Disturbance
lab.DL(True)
# send your custom command
response = lab.send_and_receive("RMULQ1", str)  
print("MulQ1:", response)

time.sleep(20)

print(lab.T1)
print(lab.T2)

# Load Disturbance
lab.DL(False)
# send your custom command
response = lab.send_and_receive("RMULQ1", str)  
print("MulQ1:", response)

lab.close() # Disconnect from TCLab
```
